```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv("merchant_fill_nonulls.csv", index_col = 0)
df.head(2)
```

```python
# 1.  ¿Cómo se distribuyen los precios de los productos? 
sns.histplot(x = "price", data = df, bins= 25)
plt.title("Distribución de precios")
plt.xlabel("Precio")
plt.ylabel("Conteo")
plt.xlim(0,50);

# cuando se nos pida conocer la distribución de una variable numérica la gráfica ideal es un histograma. 
```

```python
# con seaborn pero que nos de porcentaje, no recuento
sns.histplot(x = "price", data = df, bins= 25, stat='percent')
plt.title("Distribución de precios")
plt.xlabel("Precio")
plt.ylabel("Porcentaje")
plt.xlim(0,50);
```

```python
# con matplotlib
plt.hist(x = "price", data = df, bins= 25, edgecolor = 'black')
plt.title("Distribución de precios")
plt.xlabel("Precio")
plt.ylabel("Conteo")
plt.xlim(0,50);
```

```python
# 2. ¿Cuál es la relación entre el precio y la calificación promedio de los productos?
sns.scatterplot(x = "price", y = "mean_product_ratings_count", data = df)
plt.ylabel("Puntuación media")
plt.xlabel("Precio")
plt.title("Relacion entre el precio y sus valoraciones")
plt.xlim(0,50)
plt.ylim(0, 22000);

# Al pedirnos que busquemos la relación entre dos variables numéricas podemos hacer un scatterplot o un regplot
```

```python
# scatterplot con matplotlib

plt.scatter(x = "price", y = "mean_product_ratings_count", data = df)
plt.ylabel("Puntuación media")
plt.xlabel("Precio")
plt.title("Relacion entre el precio y sus valoraciones")
plt.xlim(0,50)
plt.ylim(0, 22000);
```

```python
sns.regplot(x = "price", y = "mean_product_ratings_count", data = df)
plt.ylabel("Puntuación media")
plt.xlabel("Precio")
plt.title("Relacion entre el precio y sus valoraciones")
plt.xlim(0,50)
plt.ylim(0, 22000);
```

```python
# 3.  ¿Cómo se distribuyen los precios de los productos en función de si se utilizan anuncios promocionales (`uses_ad_boosts`) o no?
sns.histplot(x = "price", hue = "uses_ad_boosts", data = df, multiple='stack') # este gráfico tiene las columnas apiladas
plt.ylabel("Conteo")
plt.xlabel("Precio")
plt.title('Histograma de ventas por producto y uso o no de anuncios promocionales')
plt.xlim(0,50);

# graficos que nos pidan hue los hacemos con seaborn, hue no existe en matplotlib
```

```python
sns.histplot(x = "price", hue = "uses_ad_boosts", data = df, multiple='layer') # este gráfico tiene las columnas superpuestas
plt.ylabel("Conteo")
plt.xlabel("Precio");

# graficos que nos pidan hue los hacemos con seaborn, hue no existe en matplotlib
```

```python
# 4. ¿Existe una relación entre el uso de anuncios y las unidades vendidas?
sns.histplot(x = "mean_units_sold_per_product", hue = "uses_ad_boosts", data = df, bins = 20, multiple='fill')
plt.ylabel("Conteo")
plt.xlabel("Precio")
plt.title('Histograma normalizado unidades vendidas y presencia o no de anuncios promocionales');

# con un histograma con las barras apiladas en porcentaje, es decir, permite ver la proporcion de ventas si se han usado anuncios vs si no se han usado anuncios. 
```

```python
# con un barplot podemos calcular el promedio de ventas para productos que se han anunciado vs los que no

sns.barplot(y = "mean_units_sold_per_product", x = "uses_ad_boosts", data = df, hue = 'uses_ad_boosts')
plt.ylabel("Promedio unidades vendidas")
plt.xlabel("Precio")
plt.title('Promedio ventas en funcion del uso de anuncios promocionales');


# La respuesta sería que el promedia de ventas es mayor en aquellos productos que no han sido anunciados, por lo que no parece haber un impacto
# del uso de anuncios en las ventas. 
```

```python
# 5. ¿Cuál es la relación entre el precio de envío y las unidades vendidas?
sns.scatterplot(x = "retail_price", y = "units_sold", data = df)
plt.ylabel("unidades vendidas")
plt.xlabel("precio")
plt.title('Relacion entre unidades vendidas y precio productos')
plt.xlim(0, 250)
plt.ylim(0,120000);

# nos está pidiendo comparar 2 variables numéricas, scatter o regplot
```

```python
# con matplotlib
plt.scatter(x = "retail_price", y = "units_sold", data = df)
plt.ylabel("unidades vendidas")
plt.xlabel("precio")
plt.title('Relacion entre unidades vendidas y precio productos')
plt.xlim(0, 250)
plt.ylim(0,120000);
```

```python
# REGPLOT

sns.regplot(x = "retail_price", y = "units_sold", data = df)
plt.ylabel("unidades vendidas")
plt.xlabel("precio")
plt.title('Relacion entre unidades vendidas y precio productos')
plt.xlim(0, 250)
plt.ylim(0,120000);

# No parece haber relacion entre las unidades vendidas y el precio de los productos
```

```python
# 6. ¿Cuál es la distribución de las calificaciones de los empleados (`rating_employee`) para los productos que tienen el distintivo "Buena calidad del producto" (`badge_product_quality`) y para los que no lo tienen
buenos = df[df["badge_product_quality"] == "Buena"]
excelente = df[df["badge_product_quality"] == "Excelente"]
# preparamos los dos grupos
```

```python
fig, axes = plt.subplots(nrows = 2, ncols = 1, figsize = (20, 10), sharex= True, sharey=True) 
# con sharex y sharey aseguramos que los ejes estén a la misma escala y podamos hacer comparaciones
sns.histplot(x = "rating_employee", data = buenos, ax = axes[0])
sns.histplot(x = "rating_employee", data = excelente, ax = axes[1])
axes[0].set_title('Histograma de la calificacion de los empleados productos sello calidad = Bueno')
axes[1].set_title('Histograma de la calificacion de los empleados productos sello calidad = Excelente')
axes[0].set_xlabel('Calificacion empleado')
axes[1].set_xlabel('Calificacion empleado')
axes[0].set_ylabel('Recuento')
axes[1].set_ylabel('Recuento');
```

```python
# 7. Compara diferentes distribuciones de precios de productos en función de si tienen el distintivo "Producto local" o "Buena calidad del producto".
# UNA SOLUCIÓN

# Filtrar productos con distintivo "Producto local"
productos_local = df[df['badge_local_product'] == 'Si']

# Filtrar productos con distintivo "Buena calidad del producto"
productos_calidad = df[df['badge_product_quality'] == 'Buena']

# Crear subplots
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Histograma para productos "Producto local" con matplotlib
axes[0].hist(data = productos_local, x = 'price', bins=20, color='skyblue', alpha=0.7, edgecolor = 'black')
axes[0].set_title('Distribución de Precios (Producto Local)')
axes[0].set_xlabel('Precio')
axes[0].set_ylabel('Frecuencia')

# Histograma para productos "Buena calidad del producto" con seaborn
sns.histplot(data = productos_calidad, x = 'price', bins=20, color='lightcoral', alpha=0.7, ax = axes[1])
axes[1].set_title('Distribución de Precios (Buena Calidad)')
axes[1].set_xlabel('Precio')
axes[1].set_ylabel('Frecuencia')

# Añadir título general
plt.suptitle('Comparación de Distribuciones de Precios')
plt.tight_layout()
plt.show()

```
